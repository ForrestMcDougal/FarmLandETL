```python
from bs4 import BeautifulSoup
from splinter import Browser
import requests
```

```python
url = 'https://www.landsofamerica.com/United-States/farms/'
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'html.parser')
```

```python
print(soup.body.prettify())
```

```python
div_top = soup.body.find('div', class_='contentTop')
rstats = div_top.find('div', class_='rstats')
price = rstats.find('span', class_='price').text
print(price)
size = rstats.find('span', class_='size').text
print(size)
address = div_top.find(itemprop='streetAddress').text
print(address)
locality = div_top.find(itemprop='addressLocality').text
print(locality)
address_region = div_top.find(itemprop='addressRegion').text
print(address_region)
postal_code = div_top.find(itemprop='postalCode').text
print(postal_code)
county = div_top.find('span', class_='county').text
print(county.split('- ')[1].strip())
address = div_top.find('address')
# print(address.prettify())
geo = soup.find('span', itemprop='geo')
print(geo)
latitude = address.find(itemprop='latitude')
print(latitude)
longitude = address.find(itemprop='longitude')
print(longitude)
```
