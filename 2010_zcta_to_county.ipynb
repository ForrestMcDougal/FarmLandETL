Example 1

In this example, ZCTA 85609 is highlighted in yellow below and falls wholly within county 04003, which is outlined in black.  Therefore, there is only one corresponding record in the relationship file.  The 2010 population for this relationship record is 392 (POPPT) and is equal to the 2010 Censuspopulation of 392 for ZCTA 85609 (ZPOP).  Likewise, the 2010 Census housing unit count for both the relationship record (HUPT) and the ZCTA (ZHU) is 271.  Therefore, the percentage of the total population of ZCTA 85609 represented by the relationship record (ZPOPPCT) is 100.  Although ZCTA 85609 falls entirely within County 04003, County 04003 contains many different ZCTAs.  The 2010 Census population living in County 04003 is 131,346 (COPOP) but the percentage of the total population of County 04003 that lives within ZCTA 85609 and is represented by the relationship record (COPOPPCT) is 0.3

Example 2

In this example, ZCTA 85602 is highlighted below in yellow and falls in two counties:  04003 and 04019, which are outlined in black.  Therefore, there are two corresponding records in the relationship file.  The first record describes the relationship between ZCTA 85602 and County 04003.  The 2010 Census population living within ZCTA 85602 is 9,464 (ZPOP) but only 8,533 (POPPT) also live within County 04003.  This represents 90.16% of the total population (ZPOPPCT) for ZCTA 85602.  Furthermore, out of the total population living within County 04003, 6.5% (COPOPPCT) live within ZCTA 85602.  The second record describes the relationship between ZCTA 85602 and county 04019.  The 2010 Census population living with ZCTA 85602 and in county 04019 is 931 (POPPT) and represents 9.84% of the total population (ZPOPPCT) for ZCTA 85602.  Furthermore, 0.09% of the total population within county 04019 also lives in ZCTA 85602 (COPOPPCT).


POPPT - population of zip in specific county
ZPOP - zip code population
ZHU - zip housing unit count 
HUPT - housing unit count of zip in specific county
ZOPOPPCT - percent of total zip pop for portion of zip in that county 
COPOP - population in that county
COPOPPCT - percept of total county pop wihtin that zip code
ZAREA - area of zipcode
ZAREALAND - land area of zipcode
COHU - county housing units
COAREA - area of county
COAREALAND - land area of county
